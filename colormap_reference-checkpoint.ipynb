# Pick a color from the graph.

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
cmaps = [('Perceptually Uniform Sequential', ['viridis', 'plasma', 'inferno', 'magma', 'cividis']),
         ('Sequential', ['Greys', 'Purples', 'Blues', 'Greens', 'Oranges', 'Reds','YlOrBr', 'YlOrRd', 'OrRd', 'PuRd', 'RdPu', 'BuPu', 'GnBu', 'PuBu', 'YlGnBu', 'PuBuGn', 'BuGn', 'YlGn']),
         ('Sequential (2)', ['binary', 'gist_yarg', 'gist_gray', 'gray', 'bone', 'pink','spring', 'summer', 'autumn', 'winter', 'cool', 'Wistia', 'hot', 'afmhot', 'gist_heat', 'copper']),
         ('Diverging', ['PiYG', 'PRGn', 'BrBG', 'PuOr', 'RdGy', 'RdBu', 'RdYlBu', 'RdYlGn', 'Spectral', 'coolwarm', 'bwr', 'seismic']),
         ('Cyclic', ['twilight', 'twilight_shifted', 'hsv']),
         ('Qualitative', ['Pastel1', 'Pastel2', 'Paired', 'Accent', 'Dark2', 'Set1', 'Set2', 'Set3','tab10', 'tab20', 'tab20b', 'tab20c']),
         ('Miscellaneous', ['ocean', 'gist_earth', 'terrain', 'gist_stern', 'gnuplot', 'gnuplot2', 'CMRmap', 'cubehelix', 'brg', 'gist_rainbow', 'rainbow', 'jet', 'turbo', 'nipy_spectral', 'gist_ncar'])]

gradient = np.linspace(0, 1, 256)
gradient = np.vstack((gradient, gradient))

def plot_color_gradients(cmap_category, cmap_list):
    # Create figure and adjust figure height to number of colormaps
    nrows = len(cmap_list)
    figh = 0.35 + 0.15 + (nrows + (nrows-1)*0.1)*0.22
    fig, axs = plt.subplots(nrows=nrows, figsize=(6.4, figh))
    fig.subplots_adjust(top=1-.35/figh, bottom=.15/figh, left=0.2, right=0.99)

    axs[0].set_title(cmap_category + ' colormaps', fontsize=14)

    for ax, cmap_name in zip(axs, cmap_list):
        ax.imshow(gradient, aspect='auto', cmap=cmap_name)
        ax.text(-.01, .5, cmap_name, va='center', ha='right', fontsize=10,
                transform=ax.transAxes)

    # Turn off *all* ticks & spines, not just the ones with colormaps.
    for ax in axs:
        ax.set_axis_off()

for cmap_category, cmap_list in cmaps:
    plot_color_gradients(cmap_category, cmap_list)

plt.show()
```

```python
cmap_name = 'viridis'
value = 0.9
cmap = plt.get_cmap(cmap_name)
selected_color = cmap(value)
print(f"{selected_color}")
```

```python
# Draw a plot using {selected color}.
x = np.random.rand(50)
y = np.random.rand(50)
plt.figure()
plt.scatter(x, y, color=selected_color)
plt.title(f"{cmap_name}_color")
plt.show()
```

# Accumulation of color categories

```python
# Nature
Blue_1 = r"#0ddbf5"
Blue_2 = r"#1d9bf7"
Blue_3 = r"#4292c9"
Blue_4 = r"#303cf9"
Blue_5 = r"#a0c9e5"

orange_1 = r"#f26a11"
orange_2 = r"#fe5357"
orange_3 = r"#fd7c1a"
orange_4 = r"#fe9376"

red_1 = r"#ff6b62"
red_2 = r"#a30543"

yellow_1 = r"#fcff07"
yellow_2 = r"#f7ec44"
yellow_3 = r"#ffa725"
yellow_4 = r"#fcb431"

green_1 = r"#009d73"
green_2 = r"#afdd8b"
green_3 = r"#35a153"
green_4 = r"#274753"
green_5 = r"#297270"
green_6 = r"#299d8f"

purple_1 = r"#817cb9"
purple_2 = r"#bcbddd"
purple_3 = r"#6200aa"

pink_1 = r"#cb78a6"
pink_2 = r"#f3dee0"
pink_3 = r"#f6cae5"

gray_1 = r"#979797"
```

```python
import matplotlib.pyplot as plt
from matplotlib.patches import Rectangle

fig, ax = plt.subplots(figsize=(7, 7))

def draw_color_rect(color, x, y, label):
    rect = Rectangle((x, y), 1, 1, facecolor=color, edgecolor='none')
    ax.add_patch(rect)
    ax.text(x + 0.5, y + 0.5, label, ha='center', va='center', color='white' if color in ["#0ddbf5", "#1d9bf7", "#4292c9", "#303cf9", "#009d73", "#817cb9"] else 'black')

# blue
draw_color_rect(Blue_1, 0, 0, "Blue_1")
draw_color_rect(Blue_2, 1, 0, "Blue_2")
draw_color_rect(Blue_3, 2, 0, "Blue_3")
draw_color_rect(Blue_4, 3, 0, "Blue_4")
draw_color_rect(Blue_5, 4, 0, "Blue_5")

# orange
draw_color_rect(orange_1, 0, 1, "orange_1")
draw_color_rect(orange_2, 1, 1, "orange_2")
draw_color_rect(orange_3, 2, 1, "orange_3")
draw_color_rect(orange_4, 3, 1, "orange_4")

# red
draw_color_rect(red_1, 0, 2, "red_1")
draw_color_rect(red_2, 4, 2, "red_2")
draw_color_rect(red_3, 6, 2, "red_3")

# yellow
draw_color_rect(yellow_1, 1, 2, "yellow_1")
draw_color_rect(yellow_2, 2, 2, "yellow_2")
draw_color_rect(yellow_3, 3, 2, "yellow_3")
draw_color_rect(yellow_4, 5, 2, "yellow_4")

# green
draw_color_rect(green_1, 0, 3, "green_1")
draw_color_rect(green_2, 1, 3, "green_2")
draw_color_rect(green_3, 2, 3, "green_3")
draw_color_rect(green_4, 3, 3, "green_4")
draw_color_rect(green_5, 4, 3, "green_5")
draw_color_rect(green_6, 5, 3, "green_6")

# purple
draw_color_rect(purple_1, 0, 4, "purple_1")
draw_color_rect(purple_2, 1, 4, "purple_2")
draw_color_rect(purple_3, 2, 4, "purple_3")

# pink
draw_color_rect(pink_1, 0, 5, "pink_1")
draw_color_rect(pink_2, 1, 5, "pink_2")
draw_color_rect(pink_3, 2, 5, "pink_3")

# gray
draw_color_rect(gray_1, 0, 6, "gray_1")

# Set the range of the coordinate axes and the labels
ax.set_xlim(0, 7)
ax.set_ylim(0, 7)
ax.set_axis_off()

plt.show()
```
